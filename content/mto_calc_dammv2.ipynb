```python
from decimal import Decimal, getcontext
from math import sqrt

getcontext().prec = 80

def calc_delta_a(L, sqrt_P, RES, delta_B):
    L = Decimal(L)
    sqrt_P = Decimal(sqrt_P)
    RES = Decimal(RES)
    delta_B = Decimal(delta_B)
    num = delta_B * (2 ** (2 * RES))
    den = sqrt_P * (sqrt_P + (delta_B * (2 ** (2 * RES)) / L))
    return num / den

L = Decimal(input("Liquidity (L): "))
sqrt_P = Decimal(input("sqrt(P): "))
RES = Decimal(input("RES: "))
delta_B = Decimal(input("ΔB: "))

print("ΔA =", calc_delta_a(L, sqrt_P, RES, delta_B))
```
